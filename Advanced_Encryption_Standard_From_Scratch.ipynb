```python
def byteRotate(word):
  temp=[]
  for i in range(len(word)):
    if i!=3:
      temp.append(word[i+1])
    else:
      temp.append(word[0])
  return temp
def Substitution(word):
  temp=[]
  for i in range(len(word)):
    
    r=word[i][0]
    c=word[i][1]
    r=int(r,16)
    c=int(c,16)
    temp.append(S_BOX[r][c])
  return temp
def XORop(v1,v2):
  new = [(ord(a) ^ ord(b)) for a,b in zip(v1, v2)]
  strXOR=''
  for n in new:
    strXOR=strXOR+str(n)
  return strXOR
def Rcon(word,rc):
  word[0]='{:X}'.format(int(word[0],16)^int(rc,16))
  return word
def KeyGenerationXORING(w1,w2):
  temp=[]
  for i in range(len(w1)):
    bw1=int(w1[i],16)
    bw2=int(w2[i],16)
    temp.append('{:X}'.format(bw1^bw2))

  return temp
def makeValid(word):
  for k in range(len(word)):
    if len(word[k])!=2:
      if len(word[k])==1:
        word[k]='0'+word[k]
  return word
```

```python
def InvSubstitution(word):
  temp=[]
  for i in range(len(word)):
    r=word[i][0]
    c=word[i][1]
    r=int(r,16)
    c=int(c,16)
    temp.append(inv_sbox[r][c])
  return temp

def InverseSubFunction(state):
  temp=[]
  for i in range(len(state)):
    temp.append(InvSubstitution(state[i]))
  return temp

def invLeftShift(state):
  invv=np.array(state).T.tolist()
  temp=[]
  for i in range(len(invv)):
    temp.append(np.roll(np.array(invv[i]),i).tolist())
  temp=np.array(temp).T.tolist()
  return temp

def invMixColumn(state):
  state=mixColumn(mixColumn(mixColumn(state)))
  return state

def split(key):
  k=[key[i:i+2] for i in range(0, len(key), 2)]
  key=[k[i:i+4] for i in range(0,len(k),4)]
  return key
```

```python
def decryption(state,InitialKey):
  for i in range(10,0,-1):
    print('*****************Round 10***********************')
    key=InitialKey[(i*4):(i*4)+4]
    state=addRoundKey(state,key)
    print('After Round Key: ',state)
    if i<10:
      state=invMixColumn(state)
      print('After Inverse Mix column Operation: ',state)
    state=invLeftShift(state)
    print('After Inverse Row Shift Operation: ',state)
    state=InverseSubFunction(state)
    print('After Inverse Substitution Operation: ',state)
  key=InitialKey[0:4]
  state=addRoundKey(state,key)
  print('After Last Round Key Operation: ',state)
  return state
```
